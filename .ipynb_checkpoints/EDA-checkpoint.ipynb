```python
# save as download_audio.py
import sys
from yt_dlp import YoutubeDL

def download_as_mp3(url: str, out_dir: str = "downloads_mp3", bitrate: str = "192"):
    ydl_opts = {
        "format": "bestaudio/best",
        "postprocessors": [
            {
                "key": "FFmpegExtractAudio",
                "preferredcodec": "mp3",
                "preferredquality": "192",
            },
        ],
        "ffmpeg_location": "/opt/homebrew/bin",  #  change to folder containing ffmpeg & ffprobe
    }
    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

out_dir = 'audio/'
url = 'https://www.youtube.com/watch?v=bDVX-fsUdf4'
download_as_mp3(url, out_dir)
```
